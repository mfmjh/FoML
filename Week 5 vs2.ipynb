## Introduction

This notebook is dedicated to building a model that predicts if the students in this data set might have a romantic interest. For this the k-NN (K-Nearest Neighbor) algorithm is used. 

The data were obtained in a survey of students math and portuguese language courses in secondary school. It contains a lot of interesting social, gender and study information about students. The used dataset comes from a study by P. Cortez and A. Silva (2008).

## Building the model

Since I don't have that much experience with the different models I wanted to expiriment to see which model would give the most accurat predictions. I've understood that Naive Bayes could work whenever there is little correlations between variables. Therefoire I now (before making the model) think this model will be give the best predictions. 

```python
X = df_sub[['M', 'age', 'Fedu', 'reputation', 'failures', 'No schoolsup', 'Yes higher', 'absences', 'G2']] #create the X matrix

y = df['Yes romance'] #create the y-variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) #split the data, store it into different variables

X_train.head() #show the head of the training set
```

```python
#Naive Bayes
nb = MultinomialNB() #create the model
nb = nb.fit(X_train, y_train) #fit the model X=features, y=character

#k-NN
knn = KNeighborsClassifier() #create the model (standard 5)
knn = knn.fit(X_train, y_train) #fit the model X=features, y=character

#Random Forest
rf = RandomForestClassifier() #create the model
rf = rf.fit(X_train, y_train) #fit the model X=features, y=character
```

## Evaluate the model

```python
nb_score = nb.score(X_test, y_test)
knn_score = knn.score(X_test, y_test)
rf_score = rf.score(X_test, y_test)

print(f"The accuracy for Naive Bayes: {nb_score}")
print(f"The accuracy for k-NN: {knn_score}")
print(f"The accuracy for Random Forest: {rf_score}")
```

The baseline guess with the variable "No romance" (410 / (410 + 239) is 63.1%. Unfortunately, this means our models are even less accurate.

Let's create some confusion matrices to evaluate precision and recall of all the models.

### Naive Bayes

```python
y_test_pred = nb.predict(X_test) #the predicted values
cm_nb = confusion_matrix(y_test, y_test_pred) #creates a "confusion matrix"
cm_nb
```

```python
conf_matrix = pd.DataFrame(cm_nb, index=['No romantic interest (actual)', 'Romantic interest (actual)'], columns = ['No romantic interest (predicted)', 'Romantic interest (predicted)']) 
conf_matrix
```

```python
print(classification_report(y_test, y_test_pred))
```

###  K-nearest neighbors

```python
y_test_pred = knn.predict(X_test) #the predicted values
cm_knn = confusion_matrix(y_test, y_test_pred) #creates a "confusion matrix"
cm_knn
```

```python
conf_matrix = pd.DataFrame(cm_knn, index=['No romantic interest (actual)', 'Romantic interest (actual)'], columns = ['No romantic interest (predicted)', 'Romantic interest (predicted)']) 
conf_matrix
```

```python
print(classification_report(y_test, y_test_pred))
```

###  Random Forest

```python
y_test_pred = rf.predict(X_test) #the predicted values
cm_rf = confusion_matrix(y_test, y_test_pred) #creates a "confusion matrix"
cm_rf
```

```python
conf_matrix = pd.DataFrame(cm_rf, index=['No romantic interest (actual)', 'Romantic interest (actual)'], columns = ['No romantic interest (predicted)', 'Romantic interest (predicted)']) 
conf_matrix
```

```python
print(classification_report(y_test, y_test_pred))
```

When evaluating precision and recall, precision can be interpreted as "how many of this class do I find over the whole number of elements of this class" and recall can be interpreted as "how many of this class do I find compared the total amount in this class".

When it comes to precision there is in this case little difference between the use of Naive Bayes and Random Forest. If I interpreted the recall value in the classification report right, I would say that Random Forest best succeeded in predicting the 'romantic interest' value.
